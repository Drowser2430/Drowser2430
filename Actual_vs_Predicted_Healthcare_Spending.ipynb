```python
# STEP 1: Import libraries
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error, r2_score

# STEP 2: Load sample dataset (manually created for demo)
data = {
    'Year': [2017, 2018, 2019, 2020, 2021, 2022, 2023],
    'Healthcare Spending ($B)': [700, 725, 755, 830, 910, 1000, 1100]
}
df = pd.DataFrame(data)

# STEP 3: Visualize the trend
plt.plot(df['Year'], df['Healthcare Spending ($B)'], marker='o')
plt.title('US Healthcare Spending Over Time')
plt.xlabel('Year')
plt.ylabel('Spending ($B)')
plt.grid
```

```python
# STEP 4: Train Linear Regression model
X = df[['Year']]
y = df['Healthcare Spending ($B)']

model = LinearRegression()
model.fit(X, y)

# STEP 5: Predict future spending for 2024 and 2025
future_years = pd.DataFrame({'Year': [2024, 2025]})
predictions = model.predict(future_years)

# Display predictions
for year, spend in zip(future_years['Year'], predictions):
    print(f"Predicted healthcare spending in {year}: ${spend:.2f}B")

# STEP 6: Plot predictions
plt.plot(df['Year'], y, marker='o', label='Actual')
plt.plot(future_years['Year'], predictions, marker='x', linestyle='--', color='red', label='Predicted')
plt.title('Actual vs Predicted Healthcare Spending')
plt.xlabel('Year')
plt.ylabel('Spending ($B)')
plt.legend()
plt.grid(True)
plt.show()

# STEP 7: Evaluate model (on training set)
y_pred_train = model.predict(X)
print("Mean Absolute Error:", mean_absolute_error(y, y_pred_train))
print("R^2 Score:", r2_score(y, y_pred_train))
```
